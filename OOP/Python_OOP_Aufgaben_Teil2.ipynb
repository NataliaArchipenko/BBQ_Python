### Aufgabe 2.1: Würfel
Implementiere die Klasse `Wuerfel`.
Sie soll über die Methode `werfen()` verfügen, die ein Zahl zwischen
1 und 6 im Attribut `augen` speichert.

Erzeuge ein Objekt der Klasse `Wuerfel` und würfele hiermit solange, bis der Würfel eine 6 liefert und gib die Anzahl der Versuche aus.

Für den Wurf kann folgendes benutzt werden:
```py
from random import randint
randint(1,6)
```

````python
# Dein Code
from random import randint
class Wuerfel:
    def __init__(self):
        self.augen = 0

    def werfen(self):
        self.augen = randint(1,6)
        return self.augen
    

wuerfel = Wuerfel()
anzahl_von_wurfen = 0
while True:
    anzahl_von_wurfen +=1
    ergebnis = wuerfel.werfen()
    print(f' Wurfnummer: {anzahl_von_wurfen}, das Ergebnis ist  {ergebnis}')
    if ergebnis == 6:
        break
print(f"Die Anzahl von Wurfen war {anzahl_von_wurfen}, um eine 6 zu würfeln.")
````

````python
# Dein Code
class Person:
    def __init__(self, name, alter, geschlecht):
        self.name = name
        self.alter = alter
        self.geschlecht = geschlecht

    def zeige_details(self):
        print(f'Die Name von Person ist {self.name}. Sie ist {self.alter} Jahren alt. Ihre Geschlecht ist {self.geschlecht}')
    
    def geburtstag_feiern(self):
        self.alter +=1
        return f'{self.name} hat heute der Geburtstag gefeiert. Sie/er ist  {self.alter} Jahre alt geworden'

class Student(Person):
        def __init__(self, name, alter, geschlecht, student_id, studien_fach):
            super().__init__(name, alter, geschlecht)#weil das ich von Person nehme
            self.student_id = student_id
            self.studien_fach = studien_fach
        def zeige_details(self):
             super().zeige_details()
             print(f'Student_ID  ist {self.student_id}, Studienfach ist  {self.studien_fach}')

class Angestellter(Person):
        def __init__(self, name, alter, geschlecht, mitarbeiter_id, position):
            super().__init__(name, alter, geschlecht)
            self.mitarbeiter_id = mitarbeiter_id
            self.position = position

        def zeige_details(self):
            super().zeige_details()
            print(f'Mitarbeiter_ID  ist {self.mitarbeiter_id}, Position ist  {self.position}')

class Personverwaltung(Person):
        def __init__(self):
             self.liste_von_Person = [] #leer liste

        def person_hinzufügen(self, person):
             self.liste_von_Person.append(person)
             print(f' Die Person {person.name} wurde in die Liste hinzugefügt')

        def person_entfernen(self, person):
             self.liste_von_Person.remove(person)#besser, weil man kein Index braucht, um Person zu entfernen
             print(f' Die Person {person.name} wurde entfert')

        def zeige_alle_personen(self):
            for person in self.liste_von_Person:
                person.zeige_details()
                        

#test
person1 = Person("Maria", 18, "weiblich")
person2 = Person("Konstantin", 35, 'männlich')
person1.zeige_details()
person2.zeige_details()
#person1.geburtstag_feiern()
person2.geburtstag_feiern()

student = Student("Katy", 18, "weiblich", 1, "Informatik")
student.zeige_details()

mitarbeiter = Angestellter('Alexandr Podolnikov', 43, 'männlich', 11, 'IT-Systemadministator')
mitarbeiter.zeige_details()

verwaltung = Personverwaltung()
verwaltung.person_hinzufügen(person1)
verwaltung.person_hinzufügen(person2)
verwaltung.person_hinzufügen(student)
verwaltung.person_hinzufügen(mitarbeiter)
verwaltung.zeige_alle_personen()
verwaltung.person_entfernen(person1)
verwaltung.zeige_alle_personen()
````

### Aufgabe: 2.3 Bruch-Klasse

Erstelle eine Klasse `Bruch`, die die folgenden Attribute hat:

-   `zaehler`: Der Zähler des Bruchs (Integer).
-   `nenner`: Der Nenner des Bruchs (Integer).

**Methoden**:

-   `__init__(zaehler: int, nenner: int)`: Konstruktor, der den Zähler und den Nenner initialisiert und den Bruch gegebenenfalls kürzt. Wenn der Zähler `0` ist soll ein `ValueError` ausgelöst werden.
-   `__str__()`: Gibt den Bruch als String im Format "zaehler/nenner" zurück.
-   `kuerzen()`: Kürzt den Bruch auf seine einfachste Form.
-   `erweitern(other: Bruch)`: Passt die Zähler und Nenner des aktuellen Bruchs und des übergebenen Bruchs an, sodass beide Brüche denselben Nenner haben, um die Addition korrekt durchführen zu können. Wenn das übergebene Argument kein `Bruch` ist soll ein `TypeError` ausgelöst werden.
-   `__add__(other: Bruch)`: Addiert zwei Brüche, indem die Methode `erweitern` verwendet wird, um die Brüche auf denselben Nenner zu bringen, und gibt das Ergebnis als neuen Bruch zurück.
- Bonus: -   `__repr__()`: Gibt eine detaillierte String-Repräsentation des Bruchs zurück.
- Teste die Klasse indem Du ein paar Brüche erstellst und diese addierst. Versuche auch die beiden Fehler zu erzeugen.

````python
# Dein Code
class Bruch:
    def __init__(self, zaeler: int, nenner: int):
        if nenner == 0:
            print(ValueError)
        else:
            self.zaeler = zaeler
            self.nenner = nenner
            self.kuerzen() # bei Initialisierung der Bruch man kann das schon einfacher machen_ z.b 2/4 = 1/2

    def __str__(self):
        return self.zaeler/self.nenner
    
    def kuerzen(self):
        """Kürzt den Bruch auf die einfachste Form, indem die Werte überprüft werden."""
        for i in range(2, min(abs(self.zaehler), abs(self.nenner)) + 1):
            while self.zaehler % i == 0 and self.nenner % i == 0:
                self.zaehler //= i
                self.nenner //= i

    def __repr__(self):
        return self.zaeler, self.nenner
    
    def erweitern(self, other: 'Bruch'):
        """Erweitert den aktuellen Bruch und den übergebenen Bruch auf denselben Nenner."""
        if not isinstance(other, Bruch):
            raise TypeError("Das Argument muss ein Bruch sein.")
        
        
        neuer_nenner = self.nenner * other.nenner
        neuer_zaehler_self = self.zaehler * other.nenner
        neuer_zaehler_other = other.zaehler * self.nenner
        
        return Bruch(neuer_zaehler_self, neuer_nenner), Bruch(neuer_zaehler_other, neuer_nenner)
        
    def add(self, other):
        if not isinstance(other, Bruch):
            print(TypeError)
    # Erweitern der Brüche
        erweiterter_self, erweiterter_other = self.erweitern(other)

        # Summe der Zähler
        neuer_zaehler = erweiterter_self.zaehler + erweiterter_other.zaehler
        neuer_nenner = erweiterter_self.nenner

        return Bruch(neuer_zaehler, neuer_nenner)

    def __repr__(self):
        return f"Bruch(zaehler={self.zaehler}, nenner={self.nenner})"

# Tests
if __name__ == "__main__":
    try:
        bruch1 = Bruch(1, 2)  # 1/2
        bruch2 = Bruch(1, 3)  # 1/3
        bruch3 = Bruch(2, 4)  # 2/4, sollte zu 1/2 gekürzt werden
        
        print("Brüche:")
        print(bruch1)  # Ausgabe: 1/2
        print(bruch2)  # Ausgabe: 1/3
        print(bruch3)  # Ausgabe: 1/2, nach Kürzen
        
        summe = bruch1 + bruch2  # 1/2 + 1/3
        print(f"Summe: {summe}")  # Ausgabe: 5/6

        # Versuchen, einen Bruch mit 0 als Zähler zu erstellen
        try:
            bruch_invalid = Bruch(0, 5)
        except ValueError as e:
            print(e)  # Ausgabe: Der Zähler darf nicht 0 sein.

        # Versuchen, einen Bruch mit 0 als Nenner zu erstellen
        try:
            bruch_invalid = Bruch(5, 0)
        except ValueError as e:
            print(e)  # Ausgabe: Der Nenner darf nicht 0 sein.

        # Versuchen, einen nicht-Bruch als Argument zu übergeben
        try:
            bruch1.erweitern("nicht ein Bruch")
        except TypeError as e:
            print(e)  # Ausgabe: Das Argument muss ein Bruch sein.

    except Exception as e:
        print("Ein Fehler ist aufgetreten:", e)
        



   
````
