````python
import pandas as pd
import requests

url = "https://www.tiobe.com/tiobe-index/"

aufruf = requests.get(url)

if aufruf.status_code == 200:
    tables = pd.read_html(aufruf.text)  
    erste_tabelle = tables[0]  
    print(erste_tabelle.head(10))  
else:
    print(f"Fehler beim Abrufen der Seite: {aufruf.status_code}")
````

(b) Erstellen Sie mit Matplotlib ein Balkendiagramm (oder Tortendiagramm), um die Popularität der o.g. Sprachen grafisch zu visualisieren.

````python
# code
import matplotlib.pyplot as plt
programmier_sprachen = erste_tabelle['Programming Language.1'][:10]
ratings = erste_tabelle['Ratings'][:10].str.rstrip('%').astype(float)
print("Programmiersprachen:", programmier_sprachen)
print("Ratings:", ratings)
plt.pie(ratings, labels=programmier_sprachen, shadow=True,
    autopct='%.2f%%', startangle=10, explode=(0.1,) * len(programmier_sprachen))

plt.title('Raiting der Programmiersprachen')


plt.show()
````

(c) Erstellen Sie aus Informationen aus Teil (a) mit Pandas ein Dataframe, mit 2 Spalten `language` und `rating` und mit 10 Reihen mit den oben genannten Programmiersprachen. 

````python
# code
df = pd.DataFrame({'language': programmier_sprachen,'rating': ratings})
df
````

(d) Speichern Sie dann das Dataframe als eine csv-Datei namens `tiobe.csv` und prüfen Sie, ob die Datei angelegt worden ist.

````python
# code
import os
df.to_csv('tiobe.csv', index=None)
if os.path.exists('tiobe.csv'):
    print(f"Die Datei '{'tiobe.csv'}' wurde erfolgreich erstellt.")
else:
    print(f"Die Datei '{'tiobe.csv'}' wurde nicht erstellt.")
````

(a) Schreiben Sie ein Programm mit NumPy Arrays, das 100 Würfe mit zwei Würfeln simuliert.   
(b) Die Verteilung der gewürfelten Zahlen soll alsHistogramm dargestellt werden.
<br>
<br>
<img src='histo.png' width='350px' align="center"/>
<br>
Ein Histogramm, das die Verteilung von Würfen mit zwei Würfeln anzeigt.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
np.random.seed(42) 
wurf1 = np.random.randint(1, 7, size=100)
wurf2 = np.random.randint(1, 7, size = 100)
summe = wurf1 + wurf2
print(summe)

df = pd.DataFrame({'Summe': summe})
haeufigkeiten = df['Summe'].value_counts().sort_index()
print(haeufigkeiten)

# Histogramm erstellen
plt.hist(df['Summe'], bins=np.arange(2, 13), alpha=0.5, color='g', edgecolor='black') # summe min = 2, summ max = 12
plt.title('Histogramm der Verteilung der gewürfelten Zahlen')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.xticks(np.arange(2, 13))
plt.yticks(np.arange(0, df['Summe'].value_counts().max() + 1, 1)) 
#plt.grid(axis='y')
plt.show()
````
