1. **Wahrscheinlichkeit für besmmte Trefferzahlen**

Frage: Wie hoch ist die Wahrscheinlichkeit, dass Spieler A genau 6 Treffer bei 15Würfen und gleicher Trefferquote (72%) erzielt?

Frage: Wie hoch ist die Wahrscheinlichkeit, dass Spieler B zwischen 5 und 7Treffer bei 20 Würfen und gleicher Trefferquote (48%) erzielt?(PMF benutzen und Werte addieren)

Frage: Wie hoch ist die Wahrscheinlichkeit, dass Spieler A weniger als 2 Treffer bei11 Würfen erzielt?(PMF benutzen und Werte addieren)


```python

from math import comb
#  Funktion Binomialverteilung (PMF):
def binomial_pmf(n, k, p): # n - Anzahl Wurfen ingesammt, k - Treffer  p - Treffequote
    return comb(n, k) * (p ** k) * ((1 - p) ** (n - k))



# Spieler A (k=6, n = 15, p = 0.72)
warschein_spielerA = binomial_pmf(15, 6, 0.72)
print("P(A genau 6 Treffer bei 15 Würfen):", warschein_spielerA)

# Spieler B (k zwischen 5 und 7, n = 20, p = 0.48)
print('Variant 1\n')
summe= 0
for k in range(5,8):
    summe += binomial_pmf(20,k,0.48)
print(f'Die Warscheinlichket des Spieler B ist {summe:.2f}\n')

print('Variant 2: kurzer\n')
warschei_spielerB = sum(binomial_pmf(20, k, 0.48) for k in range(5, 8))

print("P(B 5 bis 7 Treffer bei 20 Würfen):", warschei_spielerB)

# Spieler A (weniger als 2 Treffer bei 11 Würfen, p = 0.72)
warschein_spieler_A_weniger2= binomial_pmf(11, 0, 0.72) + binomial_pmf(11, 1, 0.72)
print("P(A weniger als 2 Treffer bei 11 Würfen):", warschein_spieler_A_weniger2)

print('Schleife-Variant\n')
summe1= 0
for k in range(0,2):
    summe1 += binomial_pmf(11, k, 0.72)
print(f'Die Warscheinlichkeit des Spielers A(weniger als 2 Treffer) ist {summe1}')

print('Kurzes Variant\n')
warschei_spielerA_wen_2 = sum(binomial_pmf(11, k, 0.72) for k in range(0, 2))
print(warschei_spielerA_wen_2)

#Lambda_Ausdruck
P = lambda p_x, n, k: math.comb(n,k) * p_x**k * (1 - p_x)**(n-k)
n=15
k=6
p_x= 0.72
res=P(p_x, n, k)
print(f'Die Wahrscheinlichkeit, dass Spieler A genau 6 Treffer bei 15 ')
```

3. Erwartungswert und Varianz
Frage: Was ist der Erwartungswert (Durchschnit der Treffer) für Spieler A und B bei den gegebenen Bedingungen?


E(x)=n*p, wo n - Anzahl Versuchen, 
p - Erfolgswarscheinlichkeit des Ereignisses(0.72, 0.48)

Varianz:
V(x)=n*p(1-p)= E(x)(1-p)

```python

def erwartungswert(n,p):
    return n*p

def varianz(n,p):
    return erwartungswert(n,p)*(1-p)

print(f'Die Erwatrungswert des Spilers A bei 11 Versuchen: {erwartungswert(11,0.72)}')
print(f'Die Varianz des Spilers A bei 11 Versuchen: {varianz(11, 0.72):.2f}')
print(f'Die Erwatrungswert des Spilers B bei 20 Versuchen: {erwartungswert(20, 0.48)}')
print(f'Die Varianz des Spilers B bei 20 Versuchen: {varianz(20, 0.48):.2f}\n')

print(f'Die Erwatrungswert des Spilers A bei 15 Versuchen: {erwartungswert(15,0.72):.2f}')
print(f'Die Varianz des Spilers A bei 15 Versuchen: {varianz(15, 0.72):.2f}')
print(f'Die Erwatrungswert des Spilers B bei 15 Versuchen: {erwartungswert(15, 0.48):.2f}')
print(f'Die Varianz des Spilers B bei 15 Versuchen:  {varianz(15, 0.48):.2f}')
```
