```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm

x = np.linspace(-4, 4, 50)

y = norm.pdf(x, loc=0, scale=1)

sns.kdeplot(df_norm, label="Sample Data")
plt.plot(x, y, ls="--", c="r", label="STD Normalverteilung")
plt.legend();
```
