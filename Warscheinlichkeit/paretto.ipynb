```python
import matplotlib.pyplot as plt
import numpy as np

# Daten des Beispiels
produkte = ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D', 'Produkt E']
verkaeufe = [1000, 600, 200, 100, 100]
umsatz_prozent = [50, 30, 10, 5, 5]

# Berechne die kumulierten Prozents채tze
kum_umsatz_prozent = np.cumsum(umsatz_prozent)

# Erstelle das Balkendiagramm
fig, ax1 = plt.subplots(figsize=(8, 6))

# Erstelle Balken f체r die Produkte
ax1.bar(produkte, umsatz_prozent, color='blue', alpha=0.6, label='Umsatz %')

# Erstelle eine zweite Achse f체r das kumulierte Prozent
ax2 = ax1.twinx()
ax2.plot(produkte, kum_umsatz_prozent, color='red', marker='o', label='Kumulierte %', linestyle='-', linewidth=2)

# Achsenbeschriftungen und Titel
ax1.set_xlabel('Produkte')
ax1.set_ylabel('Umsatz %', color='blue')
ax2.set_ylabel('Kumulierte %', color='red')
plt.title('Pareto-Diagramm f체r den Umsatz von Produkten')

# Zeige die Legende
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Zeige das Diagramm
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Daten des Beispiels
produkte = ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D', 'Produkt E']
verkaeufe = [1000, 600, 200, 100, 100]
umsatz_prozent = [50, 30, 10, 5, 5]

# Berechne die kumulierten Prozents채tze
kum_umsatz_prozent = np.cumsum(umsatz_prozent)

# Erstelle das Balkendiagramm
fig, ax1 = plt.subplots(figsize=(10, 6))

# Erstelle Balken f체r die Verkaufszahlen
ax1.bar(produkte, verkaeufe, color='blue', alpha=0.6, label='Verk채ufe')

# Erstelle eine zweite Achse f체r den kumulierten Umsatzprozentsatz
ax2 = ax1.twinx()
ax2.plot(produkte, kum_umsatz_prozent, color='red', marker='o', label='Kumulierte Umsatz %', linestyle='-', linewidth=2)

# Achsenbeschriftungen und Titel
ax1.set_xlabel('Produkte')
ax1.set_ylabel('Verk채ufe', color='blue')
ax2.set_ylabel('Kumulierte Umsatz %', color='red')
plt.title('Pareto-Diagramm: Verk채ufe und Umsatzverteilung')

# Zeige die Legende
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Zeige das Diagramm
plt.show()
```
