**Die Standartiesirung einfach**

Die Standardisierung wird in der Statistik verwendet, um unterschiedliche Datens√§tze vergleichbar zu machen und zu analysieren. Sie ist besonders n√ºtzlich, wenn man Daten hat, die unterschiedliche Einheiten oder Ma√üst√§be haben. Bei der Standardisierung werden die Daten so umgewandelt, dass ihr Mittelwert 0 und ihre Standardabweichung 1 ist.

Warum ist die Standardisierung wichtig?
Vergleichbarkeit: Erm√∂glicht den Vergleich von Werten aus unterschiedlichen Datens√§tzen oder mit unterschiedlichen Einheiten.
Normalisierung: Viele statistische Methoden, wie Regressionsanalysen oder maschinelles Lernen, funktionieren besser mit standardisierten Daten.
Z-Scores: Die Standardisierung hilft, Z-Scores zu berechnen, die zeigen, wie viele Standardabweichungen ein Wert vom Mittelwert entfernt ist.
Wie funktioniert die Standardisierung?
Der Standardisierungsprozess wird folgenderma√üen durchgef√ºhrt:

Mittelwert subtrahieren: Jeder Datenpunkt wird um den Mittelwert verschoben, sodass der neue Mittelwert der standardisierten Daten 0 ist.
Durch die Standardabweichung teilen: Jeder Wert wird dann durch die Standardabweichung geteilt, sodass die neue Verteilung eine Standardabweichung von 1 hat.
Formel:

ùëß = (ùë• ‚àí ùúá) / ùúé

 
z: Der standardisierte Wert (Z-Score)

x: Ein einzelner Datenpunkt

Œº: Mittelwert der Daten

œÉ: Standardabweichung der Daten

**Einfaches Beispiel:**
Stell dir vor, du hast zwei Sch√ºler, die Mathe- und Englischtests geschrieben haben, aber die Tests haben unterschiedliche Skalen:

Mathe-Test: Mittelwert = 70, Standardabweichung = 10, Sch√ºler A hat 85 Punkte.
Englisch-Test: Mittelwert = 50, Standardabweichung = 5, Sch√ºler B hat 60 Punkte.
Die Frage ist: Wer hat relativ besser abgeschnitten?

Schritt 1: Berechnung der Z-Scores:

F√ºr Sch√ºler A im Mathe-Test:
ùëß = (85 ‚àí 70)/ 10 = 1.5

F√ºr Sch√ºler B im Englisch-Test:
ùëß = (60‚àí50) / 5 =2

**Ergebnis:**

Sch√ºler A hat einen Z-Score von 1.5, das bedeutet, er liegt 1.5 Standardabweichungen √ºber dem Mittelwert des Mathe-Tests.
Sch√ºler B hat einen Z-Score von 2.0, das bedeutet, er liegt 2 Standardabweichungen √ºber dem Mittelwert des Englisch-Tests.
Interpretation: Sch√ºler B hat im Verh√§ltnis zum Durchschnitt seiner Gruppe besser abgeschnitten, obwohl die absoluten Punkte (60 vs. 85) weniger erscheinen.


***Die Standardisierung hilft, Daten vergleichbar zu machen, indem sie die Einheiten angleicht. Das ist n√ºtzlich, wenn man verstehen will, wie weit ein Datenpunkt im Verh√§ltnis zu seiner Gruppe vom Durchschnitt entfernt ist.***



```python
import numpy as np

# Testdaten
mathe_scores = [85]  # Sch√ºler A's Punktzahl im Mathe-Test
englisch_scores = [60]  # Sch√ºler B's Punktzahl im Englisch-Test

# Mittelwert und Standardabweichung der Tests
mean_mathe = 70
std_mathe = 10
mean_englisch = 50
std_englisch = 5

# Berechnung der Z-Scores
z_score_mathe = (mathe_scores[0] - mean_mathe) / std_mathe
z_score_englisch = (englisch_scores[0] - mean_englisch) / std_englisch

print(f"Z-Score Mathe: {z_score_mathe:.2f}")
print(f"Z-Score Englisch: {z_score_englisch:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Testdaten
scores = {
    'Test': ['Mathe', 'Englisch'],
    'Punktzahl': [85, 60],
    'Mittelwert': [70, 50],
    'Standardabweichung': [10, 5]
}

# Berechnung der Z-Scores
scores['Z-Score'] = [(scores['Punktzahl'][i] - scores['Mittelwert'][i]) / scores['Standardabweichung'][i] for i in range(len(scores['Test']))]

# Visualisierung der Z-Scores
sns.barplot(x=scores['Test'], y=scores['Z-Score'])
plt.axhline(0, color='gray', linestyle='--')
plt.title('Z-Scores der Tests')
plt.ylabel('Z-Score')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Parameter der Normalverteilung
mean = 0  # Mittelwert der standardisierten Normalverteilung
std_dev = 1  # Standardabweichung der standardisierten Normalverteilung

# Daten f√ºr die Normalverteilungskurve
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mean, std_dev)

# Plot der Normalverteilungskurve
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standardisierte Normalverteilung', color='blue')

# Z-Scores zur Visualisierung
z_scores = [1.5, 2.0]  # Beispiel-Z-Scores f√ºr Mathe und Englisch
for z in z_scores:
    plt.axvline(z, color='red', linestyle='--', label=f'Z-Score: {z}')

# Diagrammeinstellungen
plt.title('Normalverteilungskurve und Z-Scores')
plt.xlabel('Z-Werte')
plt.ylabel('Dichte')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Parameter f√ºr die Normalverteilungen
mean_standard = 0
std_standard = 1

mean_mathe = 70
std_mathe = 10

mean_englisch = 50
std_englisch = 5

# X-Bereich f√ºr die Kurven
x_standard = np.linspace(-4, 4, 1000)
x_mathe = np.linspace(mean_mathe - 3*std_mathe, mean_mathe + 3*std_mathe, 1000)
x_englisch = np.linspace(mean_englisch - 3*std_englisch, mean_englisch + 3*std_englisch, 1000)

# Wahrscheinlichkeitsdichtefunktionen
y_standard = norm.pdf(x_standard, mean_standard, std_standard)
y_mathe = norm.pdf(x_mathe, mean_mathe, std_mathe)
y_englisch = norm.pdf(x_englisch, mean_englisch, std_englisch)

# Plotten der Kurven
plt.figure(figsize=(12, 6))
plt.plot(x_standard, y_standard, label='Standardisierte Normalverteilung', color='blue')
plt.plot(x_mathe, y_mathe, label='Mathe-Test (Mittelwert=70, Std=10)', color='green')
plt.plot(x_englisch, y_englisch, label='Englisch-Test (Mittelwert=50, Std=5)', color='orange')

# Z-Scores zur Visualisierung
z_scores_mathe = [(85 - mean_mathe) / std_mathe]
z_scores_englisch = [(60 - mean_englisch) / std_englisch]

for z in z_scores_mathe:
    plt.axvline(mean_mathe + z * std_mathe, color='green', linestyle='--', label=f'Z-Score Mathe: {z:.2f}')

for z in z_scores_englisch:
    plt.axvline(mean_englisch + z * std_englisch, color='orange', linestyle='--', label=f'Z-Score Englisch: {z:.2f}')

# Diagrammeinstellungen
plt.title('Normalverteilungskurven f√ºr Standardisiert, Mathe und Englisch')
plt.xlabel('Werte')
plt.ylabel('Dichte')
plt.legend()
plt.grid(True)
plt.show()
```
