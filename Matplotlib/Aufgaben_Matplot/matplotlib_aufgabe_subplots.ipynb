```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import e, pi, sin, exp, cos

plt.figure(figsize=(6.4,4.8))

#SUPLOT 221: 

plt.subplot(221, facecolor='white')

plt.xticks(())
plt.yticks(())

# Array 
x = np.arange(-5.0, 1.0, 0.1)
# Überschrift
title = 'Die Funktion foo'
# Funktion
def foo(t): 
    return exp(-t) * cos(2*pi*t)

plt.plot(foo(x))
plt.title(title)

#SUPLOT 222: 

plt.subplot(222, facecolor='white')

# Array 
x = np.arange(-5.0, 1.0, 0.1)

# Überschrift
title = 'foopi, die Ableitung von foo'

# Funktion
def foopi(t):
    return -2*pi * exp(-t) * sin(2*pi*t) - e**(-t)*cos(2*pi*t)

plt.plot(foopi(x))
plt.title(title)
plt.xticks(())
plt.yticks(())

#SUPLOT 223: 

plt.subplot(223, facecolor='white')

# Array 
y = np.arange(-3.0, 2.0, 0.02)

# Überschrift
title = 'Die Funktion go'

# Funktion
def go(t):
    return sin(t) * cos(1/(t+0.1))
plt.plot(go(y))
plt.title(title)
plt.xticks(())
plt.yticks(())

plt.subplot(224, facecolor='#4F7942')
# Array 
z = np.arange(-0.2, 0.2, 0.001)

# Überschrift
title = 'Ein detaillierterer Blick auf go'

# Funktion
def g(t):
    return sin(t) * cos(1/(t))
plt.xticks(())
plt.yticks(())
plt.plot(g(z), color = 'black')
plt.tight_layout()
```
