```python
import matplotlib.pyplot as plt

# x- und y-Daten mit den gewünschten Schritten und Bereichen
x = [0, 7,20,40,50]
y = [0, 20,60,120,150]
plt.xlim(0, 51)   
plt.ylim(0, 161)
#plt.tick_params(axis='both', which='both', top='both', right='both', direction = 'in')
# Liniendiagramm erstellen
plt.plot(x, y, color='b')    # marker='o' fügt Punkte zu den x-y-Paaren hinzu
plt.title('Draw a line')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#plt.twinx()
#plt.twiny()


# Diagramm anzeigen
plt.show()

#y = 3*x bearbeiten
```

```python
# Koordinate als Listen
x = [1,2,3]
y = [2,4,1]
plt.margins(0.0,0.0)
plt.tick_params(axis='both', which='both', top='both', right='both', direction = 'in')
# plott erstellen
plt.plot(x,y, color = 'b', linewidth=0.7)

# Achsenbezeichnungen und Überschrift
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sample graph!')

# Grafik anzeigen
plt.show()
```

```python
# Koordinate als Listen
x = [10,20,30]
y = [20,40,10]
plt.margins(0.0,0.0)
plt.plot(x,y, color = 'b', linewidth=0.7, label='Line 1')
x1 = [10,20,30]
y1 =  [40,10 ,30]
plt.margins(0.0,0.0)
# plott erstellen
plt.plot(x1,y1, color = 'g', linewidth=0.7, label='Line 2')
plt.tick_params(axis='both', which='both', top='both', right='both', direction = 'in')
# Achsenbezeichnungen und Überschrift
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two or more lines on same plot with suitable legends')
plt.legend()

# Grafik anzeigen
plt.show()
```

```python
# Koordinate als Listen
x = [10,20,30]
y = [20,40,10]
plt.margins(0.0,0.0)
plt.plot(x,y, color = 'b', linewidth=3.0, label='Line 1-width-3')
x1 = [10,20,30]
y1 =  [40,10 ,30]
plt.margins(0.0,0.0)
# plott erstellen
plt.plot(x1,y1, color = 'r', linewidth=5.0, label='Line 2-width-5')
plt.tick_params(axis='both', which='both', top='both', right='both', direction = 'in')
# Achsenbezeichnungen und Überschrift
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two or more lines with different widths and colors with suitable legends')
plt.legend()

# Grafik anzeigen
plt.show()
```

```python
# Koordinate als Listen
x = [10,20,30]
y = [20,40,10]
plt.margins(0.0,0.0)
plt.plot(x,y, color = 'b', linewidth=3.0, linestyle = (0, (0.3, 1.0)), label='Line 1-dotted')
x1 = [10,20,30]
y1 =  [40,10 ,30]
plt.margins(0.0,0.0)
# plott erstellen
plt.plot(x1,y1, color = 'r', linewidth=5.0, linestyle = (0,(1,1)), label='Line 2-dashed')#1-länge der schricht, 1-länger Lucke

# Achsenbezeichnungen und Überschrift
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot with two or more lines with different styles')
plt.legend()

# Grafik anzeigen
plt.show()
```

```python
import numpy as np

x = np.array([1,4,5,6,7])
y = np.array([2,6,3,6,3])

plt.margins(0.2,0.5)
plt.scatter(x, y, color='blue', s=150, label='Blaue Punkte')  # s=100 für größere Punkte
plt.plot(x, y, color='red', linestyle=(0, (0.5,1.5,1, 1.5)), linewidth=2.5)
plt.scatter(x, y, color='blue', s=150, label='Blaue Punkte')  # s=100 für größere Punkte
plt.xlim(1, 8)  
plt.ylim(1, 8)
plt.xlabel('x-Achse') # Bezeichnung für die x-Achse
plt.ylabel('y-Achse') # Bezeichnung für die y-Achse
plt.title('Display marker')
plt.show()
```

```python
# Koordinate erstellen

x1 = [2,3,5,6,8]
y1 = [1,5,10,18,20]
plt.scatter(x1, y1, color='blue', s=30, edgecolor='black', marker='*', label='Blaue Sterne') 
x2 = [3,4,6,7,9]
y2 = [2,6,12,20,22]
plt.scatter(x2, y2, color='red', s=30, edgecolor='black', marker='o', label='Rote Punkte') 
plt.xlim(0, 10)  
plt.ylim(0, 30)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4.8, 25)

y1 = x**2           
y2 = x**3           
y3 = np.sqrt(x)     


plt.xlim(0, 5)
plt.ylim(0, 120)

plt.scatter(x, y1, marker='s', s=20, color='blue', edgecolors='black') 
plt.scatter(x, y2, marker='^',s=20, color='red', edgecolors='black')
plt.plot(x, y3, color='green', linestyle=(0,(0,3,1)), linewidth=1) 


plt.show()
```

```python
# Erstelle ein Figure-Objekt
fig = plt.figure(figsize=(10, 6))
plt.subplot(221, facecolor='black') # alternativ .subplot(2,2,1)
plt.text(
    0.5, # x-Koordinaten hier anfangen zu zeigen
    0.5, # y-Koordinate hier anfangen zu zeigen
    'subplot(2,2,1)', # Textbeschreibung
    ha = 'center', # Horizontal alignment (alternativ horizontalalignment='center')
    va = 'center', # Vertical alignment (alternativ verticalalignment='center')
    fontsize = 18, # Schriftgröße
    alpha = 0.5, # Durchsichtigkeit (zw. 0.0 und 1.0)
    color = 'red' # Textfarbe)
)
plt.subplot(224, facecolor='green') # Farbe des Hintergrunds
plt.text(
    0.5,
    0.5,
    'subplot(2,2,4)', # Textbeschreibung
    ha = 'center', # Horizontal alignment (alternativ horizontalalignment='center')
    va = 'center', # Vertical alignment (alternativ verticalalignment='center')
    fontsize = 18, # Schriftgröße
    alpha = 1.0, # Durchsichtigkeit (zw. 0.0 und 1.0)
    color = 'yellow' # Textfarbe)
)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Erstelle ein Figure-Objekt
fig = plt.figure(figsize=(6, 6))

# Erstelle ein GridSpec-Objekt
gs = gridspec.GridSpec(3, 3)

# Füge einen großen Subplot hinzu, der über mehrere Zeilen und Spalten geht
ax1 = fig.add_subplot(gs[0, :])  # Dieser geht über die gesamte obere Reihe
plt.xticks(())
plt.yticks(())

# Füge einen kleineren Subplot hinzu
ax2 = fig.add_subplot(gs[1, 0])  # Dieser ist in der ersten Spalte der zweiten Reihe
plt.xticks(())
plt.yticks(())

# Füge einen weiteren kleinen Subplot hinzu
ax3 = fig.add_subplot(gs[1, 1])  # Dieser ist in der zweiten Spalte der zweiten Reihe
plt.xticks(())
plt.yticks(())
# Füge einen weiteren Subplot hinzu
ax4 = fig.add_subplot(gs[1, 2])  # Dieser ist in der dritten Spalte der zweiten Reihe

# Passe die Layouts an
plt.tight_layout()
plt.xticks(())
plt.yticks(())
plt.show()
```
