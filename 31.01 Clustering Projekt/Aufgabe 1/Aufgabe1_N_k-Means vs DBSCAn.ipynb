### Aufgabe:
Wende die Clustering-Verfahren K-Means und DBSCAN auf die bereitgestellten Punktmengen an.
Visualisiere die Ergebnisse und vergleiche, welches Verfahren besser für welche Punktmenge geeignet ist.

### Schritte:
1. Wende K-Means auf die Punktmengen an und visualisiere die Ergebnisse.
2. Wende DBSCAN auf die Punktmengen an und visualisiere die Ergebnisse.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons, make_circles
from sklearn.cluster import KMeans, DBSCAN

#Punktenmengen: 
X1, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=42)  
X2, _ = make_moons(n_samples=300, noise=0.05, random_state=42)  # Halbmonde
X3, _ = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)  
X4, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.2, random_state=42)  
X5, _ = make_moons(n_samples=300, noise=0.15, random_state=42)  
X6, _ = make_circles(n_samples=300, factor=0.4, noise=0.1, random_state=42)  
```

```python
# K-Means für X1 
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X1)
y_kmeans = kmeans.predict(X1)
```

```python
# Visualisierung
plt.scatter(X1[:, 0], X1[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X')  # Clusterzentren
plt.title("K-Means Clustering auf X1")
plt.show()
```

```python
# DBSCAN für X1:
dbscan = DBSCAN(eps=0.3, min_samples=10)
y_dbscan = dbscan.fit_predict(X1)
```

```python
# Visualisierung
plt.scatter(X1[:, 0], X1[:, 1], c=y_dbscan, s=50, cmap='viridis')
plt.title("DBSCAN Clustering auf X1")
plt.show()
```

```python
datamenge = [X1, X2, X3, X4, X5, X6]
for i, data in enumerate(datamenge, 1):
    kmeans = KMeans(n_clusters=3, random_state=42)
    kmeans.fit(data)
    y_kmeans = kmeans.predict(data)
    
    plt.figure(figsize=(8, 6))
    plt.scatter(data[:, 0], data[:, 1], c=y_kmeans, s=50, cmap='viridis')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X')
    plt.title(f"K-Means Clustering auf X{i}")
    plt.show()


```

```python
# DBSCAN Clustering für alle Punktmengen
for i, data in enumerate(datamenge, 1):
    dbscan = DBSCAN(eps=0.3, min_samples=10)
    y_dbscan = dbscan.fit_predict(data)
    
    plt.figure(figsize=(8, 6))
    plt.scatter(data[:, 0], data[:, 1], c=y_dbscan, s=50, cmap='viridis')
    plt.title(f"DBSCAN Clustering auf X{i}")
    plt.show()
```

```python
data_test = [X1, X2, X3, X4, X5, X6]

# Erstellen des Subplots:
fig, axes = plt.subplots(6, 2, figsize=(12, 24))  # 6 Zeilen, 2 Spalten

for i, data in enumerate(data_test):
    # K-Means Clustering
    kmeans = KMeans(n_clusters=3, random_state=42)
    kmeans.fit(data)
    y_kmeans = kmeans.predict(data)
    
    ax_kmeans = axes[i, 0]  # K-Means wird in der linken Spalte angezeigt
    ax_kmeans.scatter(data[:, 0], data[:, 1], c=y_kmeans, s=50, cmap='viridis')
    ax_kmeans.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X')  
    ax_kmeans.set_title(f"K-Means Clustering auf X{i+1}")
    
    # DBSCAN Clustering
    dbscan = DBSCAN(eps=0.3, min_samples=10)
    y_dbscan = dbscan.fit_predict(data)
    
    ax_dbscan = axes[i, 1]  # DBSCAN wird in der rechten Spalte angezeigt
    ax_dbscan.scatter(data[:, 0], data[:, 1], c=y_dbscan, s=50, cmap='viridis')
    ax_dbscan.set_title(f"DBSCAN Clustering auf X{i+1}")
    

plt.tight_layout()
plt.show()
```
