Reshape
Oft ist es hilfreich, die Elemente eines Arrays anders anzuordnen. So kann man z.B. aus einem eindimensionalen Array der Länge 15, einen zweidimensionalen Array mit der Shape (3, 5) machen (drei Zeilen, fünf Spalten).

````python
arr_1d = np.arange(15)
arr_2d = arr_1d.reshape((3, 5))
````

Machen Sie aus arr_2d einen Array mit

- fünf Zeilen und drei Spalten!

- fünfzehn Zeilen und einer Spalte

- einer Zeile und fünfzehn Spalten

- einen eindimensionalen Array

````python
print('Fünf Zeilen und drei Spalten :\n' , arr_2d.reshape(5,3), end = '\n\n')

arr_2d_15z_1s = arr_2d.reshape((15,1))
print(f' Fünfzehn Zeilen und eine Spalte:\n {arr_2d_15z_1s}\n\n')

arr_2d_1z_15s = arr_2d.reshape((1,15))
print(f'Eine Zeile und fünfzehn Spalten:\n {arr_2d_1z_15s}\n\n')

arr_2d_eindim= np.arange(15)
print(arr_2d_eindim) # habe ich das richtig verstanden?

````

#### Aufgabe (4)

  
Schreibe zwei Python-Programme, welche jeweils eine Liste, und einen NumPy-Array aus hunderttausend ganzen Zahlen generieren. Diese Zahlen können gern durch Zufall generiert werden.

Dann ermitlle die Größe der Liste und die Größe des Arrays.  
Das ``sys``-Modul aus der Pythons Standardbibliothek verfügt über eine Funktion ``getsizeof``, um die Größe von Objekten zu berechnen.
Wir müssen die Größe der Datenstruktur (z.B. der Liste) und die Größe der enthaltenen Elemente addieren !

Zufallszahlen können mit Hilfe der Bibliothek `random` erzeugt werden, z.B. mit der Funktion `randomint(a,b)`. Diese erzeugt eine zufällige Ganzahl zwischen `a` und `b`.
Numpy hat eine eigene `random` Funktionen, hier kann `np.random.randint(a, b, size=anzahl)` benutzt werden.

````python
import numpy as np
import random
import sys

liste = [random.randint(0, 100000) for _ in range(100000)]
print(f'Random Liste: \n {liste}\n')


arr = np.random.randint(1, 100000, 100000)
print(f'Random-Array: \n {arr}')


liste_groesse = sys.getsizeof(liste)
print(f' Die Größe von der Liste: {liste_groesse}')

el_groesse_liste = 0
for element in liste:
    el_groesse_liste += sys.getsizeof(element)

print(f' Summe der enthaltenen Elemente der Liste ist {el_groesse_liste}')
print(f'Summe Liste: {liste_groesse + el_groesse_liste}')

#array_groesse = arr.nbytes
array_groesse = sys.getsizeof(arr) # Was ist besser?
print(f'Die Größe von der Array: {array_groesse}')

el_groesse_arr= 0
for el in arr:
    el_groesse_arr += sys.getsizeof(el)
print(f'Summe der enthaltenen Elemente des Arrays ist {el_groesse_arr}')

print(f'Summe Array: {array_groesse + el_groesse_arr}')

````

#### Aufgabe (5)

Erweitere die Programme aus Aufgabe (4), sodass die Zeitdauer bei Erstellung von Liste und Array separat berechnet werden können.  
Das Modul ``time`` aus der Pythons Standardbibliothek verfügt über eine Funktion ``time``, welche den aktuellen Zeitpunkt als eine Kommazahl zurückliefert.  

Vergleiche dann die Zeitdauer von zwei Programmen.

````python
import numpy as np
import random
import sys
import time 

start_time_list = time.time()

liste = [random.randint(0, 100000) for _ in range(100000)]
print(f'Random Liste: \n {liste}\n')

end_time_list = time.time()
time_fuer_listerstellung = end_time_list - start_time_list
print(f' Die Ersellung der Liste dauert: {time_fuer_listerstellung:.2f} Sekunden')



start_time_arr = time.time()
arr = np.random.randint(1, 100000, 100000)
print(f'Random-Array: \n {arr}')
end_time_arr = time.time()
time_fuer_erstellung_arr = end_time_arr - start_time_arr
print(f' Die Ersellung des Arrays dauert: {time_fuer_erstellung_arr:.2f}')


liste_groesse = sys.getsizeof(liste)
print(f' Die Größe von der Liste: {liste_groesse}')

el_groesse_liste = 0
for element in liste:
    el_groesse_liste += sys.getsizeof(element)

print(f' Summe der enthaltenen Elemente der Liste ist {el_groesse_liste} Sekunden')
print(f'Summe Liste: {liste_groesse + el_groesse_liste}')

#array_groesse = arr.nbytes
array_groesse = sys.getsizeof(arr) # Was ist besser?
print(f'Die Größe von der Array: {array_groesse}')

el_groesse_arr= 0
for el in arr:
    el_groesse_arr += sys.getsizeof(el)
print(f'Summe der enthaltenen Elemente des Arrays ist {el_groesse_arr}')

print(f'Summe Array: {array_groesse + el_groesse_arr}')

print(f' Die Erstellung der Liste dauert um {time_fuer_listerstellung -time_fuer_erstellung_arr:.2f} Sekunden länger')
````
