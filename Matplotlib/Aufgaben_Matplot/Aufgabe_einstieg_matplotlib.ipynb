## Einfache Datenvisualisierung 

#### Übung
Für die ersten 24 Tage eines Monats haben wir den Durchschnitt-Lufttemperaturen in Grad Celsius für alle jeweils drei Tage gemessen und in einer Liste gespeichert:

```python
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
```
Erstelle dafür mit Matplotlib ein Diagramm, welches entlang x-Achse die Tage (z.B. 1 bis 22) und entlang der y-Achse die Temperaturen angezeigt werden.

````python
import numpy as np
import matplotlib.pyplot as plt
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
tage = list(range(1, len(celsius_values) * 3, 3))
plt.scatter(tage, celsius_values, color='g', linewidth=2, label='Temperatur')
plt.title('Durchschnitt-Lufttemperatur')
plt.ylabel('Temperatur')
plt.xlabel('Tag')
plt.legend()

plt.show()
````
