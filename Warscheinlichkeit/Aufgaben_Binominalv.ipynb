```python
P = lambda p_x, n, k: math.comb(n, k) * p_x**k * (1 - p_x)**(n -k)
```

Frage 1: Wie hoch ist die Wahrscheinlichkeit, dass Spieler A genau 6 Treffer bei 15 Würfen und gleicher Trefferquote (72%) erzielt?

```python
n = 15
k = 6
p_x = 0.72
res = P(p_x, n, k)
print(f"Wahrscheinlichkeit, dass Spieler A genau 6 Treffer bei 15 Würfen mit einer Trefferquote 72% erzielt: {res*100:.2f} %")
```

Frage 2: Wie hoch ist die Wahrscheinlichkeit, dass Spieler B zwischen 5 und 7 Treffer bei 20 Würfen und gleicher Trefferquote (48%) erzielt?

```python
n = 20
k_r = range(5, 8)
p_x = 0.48

res = 0
for k in k_r:
    res += P(p_x, n, k)
print(f"Wahrscheinlichkeit, dass Spieler B zwischen 5 und 7 Treffer bei 20 Würfen mit einer Trefferquote von 48% erzielt: {res*100:.2f} %")
```

Frage 3: Wie hoch ist die Wahrscheinlichkeit, dass Spieler A weniger als 2 Treffer bei 11 Würfen erzielt?

```python
n = 11
k_r = range(2) 
p_x = 0.72

res = 0
for k in k_r:
    res += P(p_x, n, k)
print(f"Wahrscheinlichkeit, dass Spieler A weniger als 2 Treffer bei 1 Würfen mit einer Trefferquote von 72% erzielt: {res*100:.4f} %")
```

Bei welchem Spieler ist es wahrscheinlicher, genau 7 Treffer zu erzielen: Spieler A oder Spieler B?

```python
n = 11
k = 7
p_a = 0.72
p_b = 0.48

res_a = P(p_a,n,k)
res_b = P(p_b,n,k)

print(f"Wahrscheinlichkeit, dass Spieler A genau {k} Treffer bei {n} Würfen mit einer Trefferquote {p_a*100}% erzielt: {res_a*100:.2f} %")
print(f"Wahrscheinlichkeit, dass Spieler B genau {k} Treffer bei 15 Würfen mit einer Trefferquote {p_b*100}% erzielt: {res_b*100:.2f} %")
```

```python
import matplotlib.pyplot as plt

# Varieren der Anzahl der Würfe (n) von 7 bis 20
n_values = range(6, 21)
res_a = []
res_b = []
k = 7

# Berechnung der Wahrscheinlichkeiten für jeden Wert von n
for n in n_values:
    res_a.append(P(p_a, n, k))  # Wahrscheinlichkeit für Spieler A
    res_b.append(P(p_b, n, k))  # Wahrscheinlichkeit für Spieler B

# Plotten der Ergebnisse
plt.plot(n_values, res_a, label="Spieler A (p = 0.72)", marker='o')
plt.plot(n_values, res_b, label="Spieler B (p = 0.48)", marker='x')
plt.xlabel("Anzahl der Würfe (n)")
plt.ylabel(f"Wahrscheinlichkeit, genau {k} Treffer zu erzielen")
plt.title(f"Wahrscheinlichkeit für {k} Treffer bei verschiedenen Würfen")
plt.legend()
plt.grid(True)
plt.show()
```

 Wie verändert sich die Wahrscheinlichkeit, dass Spieler A 6 Treffer erzielt, wenn wir die Anzahl der Würfe von 11 auf 15 bei gleicher Trefferwahrscheinlichkeit erhöhen?

```python
n1 = 11
n2 = 15
k = 6
p_a = 0.72

w11 = P(p_a, n1, k)
w15 = P(p_a, n2, k)

print(f"Die Wahrscheinlichkeit sinkt von {w11*100:.2f} % zu {w15*100:.2f} %")
```

Was ist der Erwartungswert (Durchschnitt der Treffer) für Spieler A und B bei den gegebenen Bedingungen?

```python
na = 11
p_a = 0.72

nb = 15
p_b = 0.48

ea = na * p_a
eb = nb * p_b

print(f"Erwartungswert für Spieler A: bei 11 Versuchen werden im Durchschnitt {ea:.2f} Treffer erzielt")
print(f"Erwartungswert für Spieler B: bei 15 Versuchen werden im Durchschnitt {eb:.2f} Treffer erzielt")
```

Wie groß ist die Varianz der Trefferzahl für Spieler A und B? Was sagt diese Varianz über die Streuung der Ergebnisse aus?

```python
vara = na * p_a * (1 - p_a)

varb = nb * p_b * (1 - p_b)

print(f"Varianz für Spieler A: {vara:.2f}, Standardabweichung: {vara**.5:.2f}")
print(f"Varianz für Spieler B: {varb:.2f}, Standardabweichung: {varb**.5:.2f}")
```

Die Varianz beschreibt, wie stark die Ergebnisse (Trefferzahlen) um den Erwartungswert streuen. Eine höhere Varianz bedeutet eine größere Streuung der Trefferzahlen. In diesem Fall hat Spieler B eine höhere Varianz als Spieler A, was darauf hindeutet, dass die Trefferzahlen von Spieler B tendenziell stärker schwanken.
