### Übung 1: Einfaches Liniendiagramm
Aufgabe: Erstelle ein einfaches Liniendiagramm, das die Entwicklung der Temperaturen über eine Woche darstellt. Verwende die folgenden Daten:

```py
Tage: ['Montag', 'Dienstag', 'Mittwoch', 'Donnerstag', 'Freitag', 'Samstag', 'Sonntag']
Temperaturen: [20, 22, 21, 23, 24, 26, 25]
```
Zusatz: Füge Achsentitel und eine Legende hinzu.

````python
## 
temperatur = [20, 22, 21, 23, 24, 26, 25]
tage = ['Montag', 'Dienstag', 'Mittwoch', 'Donnerstag', 'Freitag', 'Samstag', 'Sonntag']

plt.plot(tage, temperatur, color='r', linewidth=2, label='Temperatur')
plt.title('Einfaches Liniendiagramm')
plt.ylabel('Temperatur')
plt.xlabel('Tag')
plt.legend()

plt.show()

````

### Übung 4: Scatterplot erstellen
Aufgabe: Erstelle einen Scatterplot, um die Beziehung zwischen der Körpergröße und dem Gewicht von 10 Personen zu visualisieren. Verwende die folgenden Daten:
```py
Körpergröße (in cm): [160, 165, 170, 175, 180, 185, 190, 195, 200, 205]
Gewicht (in kg): [50, 60, 65, 70, 80, 85, 90, 95, 100, 110]
```
Zusatz: Füge Achsentitel und eine Trendlinie hinzu.

````python
## Dein Code 
x = np.array([160, 165, 170, 175, 180, 185, 190, 195, 200, 205]) #Körpergröße (in cm):
y = np.array([50, 60, 65, 70, 80, 85, 90, 95, 100, 110])#Gewicht (in kg): 
f = lambda x,y: y/x
y1 = f(x,y)

plt.scatter(x,y1, alpha = 1, color = 'r', marker = 'o')

# Berechnung der Trendlinie
m, b = np.polyfit(x, y1, 1)  # lineare Regression (Grad 1)
trendline = m * x + b  # Trendlinie berechnen
plt.plot(x, trendline, color='blue', linestyle ='--', label='Trendlinie') 

plt.title('Scatterplot')
plt.ylabel('Gewicht')
plt.xlabel('Körpergröße')
plt.legend()
plt.grid()

plt.show()

````

````python
import matplotlib.pyplot as plt

# Beispiel-Daten
data1 = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
data2 = [15.4, 16.7, 18.1, 19.2, 20.5, 21.8, 23.0, 24.2]
data3 = [10.1, 12.3, 14.6, 16.2, 17.5, 19.1, 20.6, 22.3]
data4 = [5.2, 6.8, 8.1, 9.4, 10.6, 11.8, 13.1, 14.5]
labels = ['A', 'B', 'C', 'D']  # Beschriftungen für das Kreisdiagramm

# X-Achse: Tage
days = list(range(1, len(data1) * 3, 3))

# Figur und Achsen vorbereiten (2x2 Raster)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Kreisdiagramm
axs[0, 0].pie([sum(data1), sum(data2), sum(data3), sum(data4)], labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'purple'])
axs[0, 0].set_title("Kreisdiagramm")

# Liniendiagramm
axs[0, 1].plot(days, data1, marker='o', color='b')
axs[0, 1].set_title("Liniendiagramm")
axs[0, 1].set_xlabel("Tage")
axs[0, 1].set_ylabel("Temperatur (°C)")
axs[0, 1].grid(True)

# Balkendiagramm
axs[1, 0].bar(days, data2, color='orange')
axs[1, 0].set_title("Balkendiagramm")
axs[1, 0].set_xlabel("Tage")
axs[1, 0].set_ylabel("Temperatur (°C)")
axs[1, 0].grid(True)

# Streudiagramm
axs[1, 1].scatter(days, data3, color='green')
axs[1, 1].set_title("Streudiagramm")
axs[1, 1].set_xlabel("Tage")
axs[1, 1].set_ylabel("Temperatur (°C)")
axs[1, 1].grid(True)

# Abstand zwischen den Diagrammen anpassen
plt.tight_layout()

# Diagramme anzeigen
plt.show()
````
