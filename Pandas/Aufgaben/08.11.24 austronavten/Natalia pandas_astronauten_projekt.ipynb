13. Visualisieren Sie die Häufigkeiten der `Space Flights`

```python
flight_counts = df['Space Flights'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
flight_counts.plot(kind='bar', color='green')
plt.title('Häufigkeiten der Space Flights')
plt.xlabel('Anzahl der Space Flights')
plt.ylabel('Häufigkeit')
plt.xticks(rotation=0)  
plt.tight_layout()  # Diagramm zeigen
plt.show()
```

16. Erstellen Sie eine Korrelationsmatriz für das (Haupt-)Dataframe

```python
numerical_df = df.select_dtypes(include=[int, float])
print(numerical_df.shape)



# Berechnung der Korrelationsmatrix
correlation_matrix = numerical_df.corr()
print(correlation_matrix)
type(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Korrelationsmatrix Heatmap")
plt.show()
```
